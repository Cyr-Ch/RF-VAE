```python
%load_ext autoreload
%autoreload 2
import numpy as np
import torch
from torchvision import datasets, transforms
from vae import *
import utils
```

```python
batch_size = 128
dataset = datasets.MNIST 

train_loader = torch.utils.data.DataLoader(
        dataset('data', train=True, download=True, transform=transforms.ToTensor()),
        batch_size=batch_size, shuffle=True,)
test_loader = torch.utils.data.DataLoader(
        dataset('data', train=False, download=True, transform=transforms.ToTensor()),
        batch_size=1000, shuffle=True,)

# Visualize a few random samples from the dataset
utils.visualize_mnist(train_loader)
```

```python
obs_dim = 784
latent_dim = 32
hidden_dims = [600, 400]  

vae = VAE(obs_dim, latent_dim, hidden_dims)
opt = torch.optim.Adam(vae.parameters(), lr=1e-3)
```

```python
max_epochs = 10
display_step = 200
for epoch in range(max_epochs):
    print(f'Epoch {epoch}')
    for ix, batch in enumerate(train_loader):
        x, y = batch
        x = x.view(x.shape[0], obs_dim)
        opt.zero_grad()
        loss = -vae.elbo(x).mean(-1)
        loss.backward()
        opt.step()
        
        if ix % display_step == 0:
            print(f'  loss = {loss.item():.2f}')
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
x, y = next(iter(test_loader))
x = x.view(x.shape[0], obs_dim)
latent = vae.encoder(x)[0]
latent_embedded = TSNE(init='pca', n_components=2).fit_transform(latent.detach().numpy())
N = 10
plt.figure(figsize=(8, 6))
plt.scatter(
        latent_embedded[:, 0],
        latent_embedded[:, 1],
        c=y,
       marker='o', edgecolor='none')
plt.colorbar(ticks=range(N))
plt.grid(True)
plt.show()
```
